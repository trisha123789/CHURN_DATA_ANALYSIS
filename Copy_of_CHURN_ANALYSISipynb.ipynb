CONVERTING THE VALUE OF 1 AS YES AND VALUE OF 0 AS NO FOR SENIOUR CITIZEN COLUMN

```python
#function
def convert(value):
  if value == 1:
      return "yes"
  else:
      return "no"
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].apply(convert)
```

VISUALIZING THE CHURN COLUMN AND FINDING OUT THE COUNT USING COUNTPLOT

```python
sns.countplot(x='Churn',data = df)
plt.title('count of customers by churn')
plt.show()
```

USING GROUP BY AND AGGREGATE FOR FINDING COUNT OF CHURN

```python
gb = df.groupby('Churn').agg({'Churn' : "count"})
print(gb)
```

```python
plt.figure(figsize = (4,4))
plt.pie(gb['Churn'],labels=gb.index,autopct ="%1.2f%%")
plt.title("Percentge Of Churn Customers")
plt.show()
```

```python
#FROM THE PIE CHART WE CAN CONCLUDE THAT 26.54% PEOPLE HAVE CHURNED OUT
#LETS EXPLORE THE REASON BEHIND IT
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="gender",data = df,hue='Churn')
plt.title("CHURN BY GENDER")
plt.show()
```

```python
plt.figure(figsize = (4,4))
sns.countplot(x ='SeniorCitizen',data =  df,hue = 'Churn')
plt.title("CHURN BY SENIOR CITIZEN")
plt.show()
```

```python
crosstab = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to percentage of total per SeniorCitizen group
crosstab_perc = crosstab.div(crosstab.sum(axis=1), axis=0) * 100

# Plot stacked bar
ax = crosstab_perc.plot(
    kind='bar',
    stacked=True,
    figsize=(4,4),
    colormap="Set2"
)

# Add percentage labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt="%.1f%%", label_type="center")

plt.title("CHURN BY SENIOR CITIZEN (Percentage)")
plt.ylabel("Percentage")
plt.xlabel("SeniorCitizen")
plt.legend(title="Churn")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up subplots: adjust rows/cols as needed
n_cols = 3
n_rows = (len(cols) + n_cols - 1) // n_cols  # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(cols):
    sns.countplot(x=col, data=df, hue='Churn', ax=axes[i])
    axes[i].set_title(f"Churn by {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Count")
    axes[i].tick_params(axis='x', rotation=20)  # rotate labels for readability

# Remove any empty subplots if len(cols) not multiple of n_cols
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Customers with fiber optic internet, no online security, no backup, or no tech support show the highest churn rates. In contrast, those with DSL or additional protective services churn less. Overall, value-added services like security and support play a strong role in reducing churn and retaining customers.
